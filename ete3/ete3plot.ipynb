# Def Fun

```python
def unroot_FastTree_tree(newick_output_file):
    rooted_tree = dendropy.Tree.get_from_path(newick_output_file, "newick")
    rooted_tree.reroot_at_node(rooted_tree.find_node_with_taxon_label('root'), update_bipartitions=False)
    tree = dendropy.Tree.get_from_string(rooted_tree.as_string('newick', suppress_rooting=False).replace('root', ''), 'newick')
    tree.ladderize(ascending=False)
    return tree
    
```

```python
#HCT116
generation_map={"A":"1","B":"11","C":'111','D':'112','E':'12','F':'121','G':'122',"H":'2'}
def generation(well,generation_map):
    row= well[9]
    return generation_map[row]
```

## 2 stylish ploting

```python
x=unroot_FastTree_tree("hct116r1.500i.nw")
t=Tree(x.as_string(schema='newick')[5:])

for leaf in t:
    try:
        leaf.name = generation(leaf.name, generation_map)
#         leaf.name = generation(leaf.name, generation_map)+":"+leaf.name+":"+str(loci_sum[leaf.name])

    except:
        continue
circular_style = TreeStyle()
circular_style.mode = "c" # draw tree in circular mode
circular_style.scale = 20
t.render("circle.tree.png", w=183, units="mm", tree_style=circular_style)
```

```python
def nwtree_plot(nw):
    x=unroot_FastTree_tree(nw)
    t=Tree(x.as_string(schema='newick')[5:])
    
    for leaf in t:
        try:
            leaf.name = generation(leaf.name, generation_map)
    #         leaf.name = generation(leaf.name, generation_map)+":"+leaf.name+":"+str(loci_sum[leaf.name])
    
        except:
            continue
    ts = TreeStyle()
    ts.show_leaf_name = True
    ts.rotation = 90
    ts.scale =  240 # 120 pixels per branch length unit
    ts.branch_vertical_margin = 5 # 10 pixels between adjacent branches
    
    # ts.show_branch_length = True
    # ts.show_branch_support = True
    # t.show(tree_style=ts)
    t.render(nw+".png", w=183, units="mm", dpi= 600,tree_style=ts)
```
