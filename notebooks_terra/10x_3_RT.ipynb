```python
def bgen(length,maxstretch=3,distance=3,):
    candi=[]
    for i in range(30000):
        x=str(RandomDNA(length))
        if 45<GC(x)<55:
            candi.append(x)
    candi=barcode.filter_stretches(candi, maxstretch)
    candi=barcode.filter_distance(candi, distance)
    return candi
```
